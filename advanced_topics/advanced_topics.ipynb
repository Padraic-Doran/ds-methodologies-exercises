```python
import pandas as pd
import numpy as np
```

```python
attendance = pd.read_csv('untidy_data/attendance.csv')
# attendance.columns
attendance = attendance.rename(columns = {'Unnamed: 0': 'Student'})
attendance = attendance.melt(id_vars=['Student'], var_name='Date', value_name='attendance_score')
# attendance = attendance.set_index('Student')
attendance
```

```python
attendance.attendance_score = np.where(attendance.attendance_score == 'P',1,np.where(attendance.attendance_score == 'A',0,np.where(attendance.attendance_score == 'T',.1,.5)))
attendance
```

```python
df = attendance.groupby(['Student']).mean()
```

```python
df = df.rename(columns = {'attendance_score': 'score'})
df.reset_index(inplace = True)
df
```
