```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from vega_datasets import data
df = data.sf_temps()
df
```

```python
df.date = pd.to_datetime(df.date)
df = df.set_index('date').sort_index()
```

```python
df.resample('D').mean().plot()
```

```python
df.resample('D').min().plot()
```

```python
df.resample('D').max().plot()
```

```python
df.resample('M').mean().idxmin().temp.month
```

```python
df.resample('M').mean().idxmax().temp.month
```

```python
extremes = df.resample('D').agg(['min', 'max'])

extremes['diff'] = extremes['temp']['max'] - extremes['temp']['min']
extremes.resample('M')['diff'].mean().idxmax().month
```

```python
plt.plot(extremes.resample('D').mean())
plt.plot(extremes.resample('D').min())
plt.plot(extremes.resample('D').max())
plt.show()
```

```python
from vega_datasets import data
seattle = data.seattle_weather()
```

```python
seattle = seattle.set_index('date').sort_index()
```

```python
seattle.resample('M').precipitation.sum().idxmax()
```

```python
seattle.resample('M').precipitation.plot()
```

```python
seattle.resample('W').wind.mean().plot()
```

```python
seattle['sunny'] = seattle['weather'] == 'sun'
seattle.resample('Y').sunny.sum().idxmax()
```

```python
seattle.groupby(seattle.index.month).precipitation.sum().idxmax()

seattle['rained'] = seattle.precipitation > 0
seattle.groupby(seattle.index.month).rained.sum().idxmax()
```

Iowa Electricity

```python
from vega_datasets import data
df = data.iowa_electricity()
```

```python
df = df.set_index('year').sort_index(axis =1)
```

```python
df = df.sort_values(by = 'year')
df['year_total'] = df.groupby('year').sum()
df['source_pct'] = df['net_generation']/df['year_total']
df
```

```python
sns.lineplot(x=df.index, y = df.net_generation, hue = df.source, data = df)
```

```python
df.pivot_table('net_generation', 'year', 'source' )
```

```python
sns.lineplot(x=df.index, y = df.net_generation, hue = df.source, data = df)
plt.show()
sns.lineplot(x=df.index, y= df.source_pct, hue = df.source, data = df)
plt.show()
```
